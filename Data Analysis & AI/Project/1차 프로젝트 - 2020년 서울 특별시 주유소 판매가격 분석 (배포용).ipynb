# 1. 데이터 불러오기

```python
# 상반기 주유소 판매 데이터 불러오기
first = pd.read_csv('../data/2020년 상반기 주유소 판매가격.csv',  encoding='cp949')
first
```

```python
second = pd.read_csv('../data/2020년 하반기 주유소 판매가격.csv',  encoding='cp949')
second
```

```python
# 상반기 판매가격 데이터 프레임 결측치 조회

first.isnull().sum()
```

```python
# 하반기 판매가격 데이터 프레임 결측치 조회

second.isnull().sum()
```

# 2. 데이터 결합

```python
# 상반기와 하반기 데이터를 상하로 결합
df = pd.concat([first,second])

df
```

# 3. 데이터 가공 및 분석

```python
# 데이터 프레임 정보 조회

df.info()
```

```python
# 상표 컬럼의 고유값 조회

df['상표'].unique()
```

```python
# 지역 컬럼의 고유값 조회

df['지역'].unique()
```

```python
# 지역 컬럼의 값을 이용하여 시와 구를 분리
df['시'] = df['지역'].str.split(" ").str[0]
df['구'] = df['지역'].str.split(" ").str[1]
df.head()
```

```python
# 시 컬럼의 서울을 서울특별시로 변환
df['시'] = df['시'].replace('서울','서울특별시')
df.head()
```

```python
# 휘발유 가격이 비싼 5개 판매가격을 출력하세요.
df.sort_values('휘발유',ascending=False).head()
```

```python
# 휘발유 가격이 저렴한 5개 판매가격을 출력하세요. (단 휘발유가 0원인 데이터는 제외)
df.sort_values('휘발유',ascending=False).dropna().tail()
```

```python
# 상표별 평균 휘발유 가격을 계산하세요.
df.pivot_table('휘발유',index='상표', aggfunc='mean')
```

```python
# 상표별 평균 휘발유 가격을 시각화하세요.
df.pivot_table('휘발유',index='상표', aggfunc='mean').plot(kind="bar")
```

```python
# 상표와 셀프여부별 평균 휘발유 가격을 계산하세요.
df.pivot_table('휘발유',index=['상표','셀프여부'], aggfunc='mean')
```
